```python
import numpy as np
from numpy_vectorized_string_filtering import numpy_isin
import pandas as pd
from faker import Faker
```

```python
faker = Faker()
Faker.seed(0)
```

```python
ndata = 10

def create_filtered_column_with_numpy(
    test_df = pd.DataFrame(
        {
            'description': [f'description {i}' for i in range(ndata)]*5
        }
    )
):
    test_df['filtered_description'] = numpy_isin(
        test_df['description'].values,
        [
            'description 0',
            'description 3',
            'description 9', 
        ]
    )
    return test_df

def create_filtered_column_with_pandas(
    test_df = pd.DataFrame(
        {
            'description': [f'description {i}' for i in range(ndata)]*5
        }
    )
):
    test_df['filtered_description'] = test_df['description']
    test_df.loc[
        ~test_df['description'].isin(
            [
                'description 0',
                'description 3',
                'description 9', 
            ]
        ),
        'filtered_description'
    ] = np.nan
    return test_df
```

```python
create_filtered_column_with_numpy()
```

```python
create_filtered_column_with_pandas()
```

```python
import timeit
setup1 = """
import pandas as pd
from numpy_vectorized_string_filtering import numpy_isin
from __main__ import create_filtered_column_with_numpy
ndata = 10
"""
setup2 = """
import pandas as pd
from __main__ import create_filtered_column_with_pandas
ndata = 10
"""
print(timeit.timeit("create_filtered_column_with_numpy()", setup=setup1, number=100))
print(timeit.timeit("create_filtered_column_with_pandas()", setup=setup2, number=100))
```
